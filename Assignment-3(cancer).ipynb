```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sb
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_excel("cancer.xlsx")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop("id",axis=1,inplace=True)
```

```python
df.head()
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df.diagnosis=le.fit_transform(df.diagnosis)
```

```python
df.head()
```

# Split X,Y

```python
x=df.drop("diagnosis",axis=1)
x.head()
```

```python
y=df[["diagnosis"]]
y.head()
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
xtrain,xtest,ytrain,ytest=tts(x,y,test_size=0.3,random_state=2)
```

```python
xtrain.head()
```

# Applying Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(xtrain,ytrain)
```

```python
pred=clf.predict(xtest)
```

# Accuracy,MSE,MAE(Using Decision Tree)

```python
from sklearn.metrics import accuracy_score,mean_absolute_error,mean_squared_error
```

```python
accuracy=accuracy_score(ytest,pred)
print("Accuracy of Decision Tree:",accuracy)
```

```python
mae=mean_absolute_error(ytest,pred)
```

```python
print("MAE of Decision Tree:",mae)
```

```python
mse=mean_squared_error(ytest,pred)
```

```python
print("MSE of Decision Tree:",mse)
```

# Applying K-NN Classification

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kc=KNeighborsClassifier()
```

```python
kc.fit(xtrain,ytrain)
```

```python
xtrain.head()
```

```python
kp=kc.predict(xtest)
```

# Evalutae Model(K-NN Classification)

```python
from sklearn.metrics import accuracy_score,classification_report,RocCurveDisplay,confusion_matrix
```

```python
print(classification_report(ytest,kp))
```

```python
accuracy1=accuracy_score(ytest,kp)
print("Accuracy of K-NN classification:",accuracy1)
```

```python
mae1=mean_absolute_error(ytest,kp)
print("MAE of K-NN classification:",mae1)
```

```python
mse1=mean_squared_error(ytest,kp)
print("MSE of K-NN classification:",mse1)
```

```python
com=confusion_matrix(ytest,kp)
com
```

```python
plt.figure(figsize=(12,8))
sb.heatmap(com,annot=True)
plt.xlabel("Predicted Result")
plt.ylabel("Actual value")
```

```python
RocCurveDisplay.from_estimator(kc,xtest,ytest)
```
