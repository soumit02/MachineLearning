# Synthetic Minority Oversampling

```python
from imblearn.combine import SMOTETomek
```

```python
sm=SMOTETomek(random_state=42)
```

```python
y.value_counts()
```

```python
x1,y1=sm.fit_resample(x,y)
```

```python
y1.value_counts()
```

# Over sampling

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ov=RandomOverSampler(random_state=42)
```

```python
y.value_counts()
```

```python
x2,y2=ov.fit_resample(x,y)
```

```python
y2.value_counts()
```

# Near Miss Under Sampling

```python
from imblearn.under_sampling import NearMiss
```

```python
nm=NearMiss()
```

```python
x3,y3=nm.fit_resample(x,y)
```

```python
y3.value_counts()
```

# K-NN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn=KNeighborsClassifier()
```

# Without Sampling

```python
from sklearn.model_selection import train_test_split as tts
```

```python
xtrain,xtest,ytrain,ytest=tts(x,y,test_size=0.3,random_state=42)
```

```python
kn.fit(xtrain,ytrain)
```

```python
xtrain.head()
```

```python
withouts_accuracy=kn.score(xtest,ytest)
print("Without Sampling Accouracy:",withouts_accuracy)
```

# Hold Out Cross Validation

```python
xtrain1,xtest1,ytrain1,ytest1=tts(x1,y1,random_state=42)
```

```python
kn.fit(xtrain1,ytrain1)
```

```python
withsam_accuracy=kn.score(xtest1,ytest1)
print("With Sampling Accouracy:",withsam_accuracy)
```

# K Fold Cross Validation

```python
from sklearn.model_selection import KFold,cross_val_score,cross_validate
```

```python
kf=KFold(n_splits=4)
```

```python
kf_accuracy=cross_val_score(kn,x1,y1,cv=kf)
print("K Fold Accouracy:",kf_accuracy)
```

```python
kf_accuracy.mean()
```

# Stratified K Fold Cross Validation

```python
from sklearn.model_selection import StratifiedKFold
```

```python
st=StratifiedKFold(n_splits=4)
```

```python
Skf_accuracy=cross_val_score(kn,x1,y1,cv=st)
print("Stratified K Fold Accouracy:",Skf_accuracy)
```

```python
Skf_accuracy.mean()
```

```python
data={"Name":['Without Sampling Accouracy','With Sampling Accouracy(Hold)','K Fold Accouracy','Stratified K Fold Accouracy']
     ,'Value':[withouts_accuracy,withsam_accuracy,kf_accuracy.mean(),Skf_accuracy.mean()]}
after_df=pd.DataFrame(data=data)
```

```python
after_df.head()
```

```python
sns.barplot(x='Value',y='Name',data=after_df)
```

```python
le=LeaveOneOut()
```
